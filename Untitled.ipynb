```python
import numpy as np
import matplotlib.pyplot as plt

def bowl_function(x):
    return x[0]**2 + x[1]**2

def bowl_function_gradient(x):
    return np.array([2 * x[0], 2 * x[1]])

def adam_optimizer_2d(f, grad_f, x_init, learning_rate=0.1, beta1=0.9, beta2=0.999, epsilon=1e-8, num_iterations=50):
    x = np.array(x_init, dtype=float)
    m = np.zeros_like(x)
    v = np.zeros_like(x)
    path = [x.copy()]

    for t in range(num_iterations):
        grad = grad_f(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * grad**2
        m_hat = m / (1 - beta1**(t + 1))
        v_hat = v / (1 - beta2**(t + 1))
        x -= learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
        path.append(x.copy())

    return np.array(path)

x_init = [1.5, 2.5]
path = adam_optimizer_2d(bowl_function, bowl_function_gradient, x_init)

x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = bowl_function(np.array([X1, X2]))

plt.contour(X1, X2, Z, levels=50, cmap='viridis')
plt.plot(path[:, 0], path[:, 1], 'r-o')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Adam Optimizer in 2D')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def bowl_function_3d(x):
    return x[0]**2 + x[1]**2 + x[2]**2

def bowl_function_3d_gradient(x):
    return np.array([2 * x[0], 2 * x[1], 2 * x[2]])

def adam_optimizer_3d(f, grad_f, x_init, learning_rate=0.1, beta1=0.9, beta2=0.999, epsilon=1e-8, num_iterations=50):
    x = np.array(x_init, dtype=float)
    m = np.zeros_like(x)
    v = np.zeros_like(x)
    path = [x.copy()]

    for t in range(num_iterations):
        grad = grad_f(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * grad**2
        m_hat = m / (1 - beta1**(t + 1))
        v_hat = v / (1 - beta2**(t + 1))
        x -= learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
        path.append(x.copy())

    return np.array(path)

x_init = [1.5, 2.5, 3.5]
path = adam_optimizer_3d(bowl_function_3d, bowl_function_3d_gradient, x_init)

x1 = np.linspace(-3, 3, 50)
x2 = np.linspace(-3, 3, 50)
x3 = np.linspace(-3, 3, 50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(path[:, 0], path[:, 1], path[:, 2], c='r', marker='o')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.set_title('Adam Optimizer in 3D')
plt.show()
```
