```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def cost_function(x, y):
    return x**2 + y**2
```

```python
def gradient(x, y):
    return 2*x, 2*y
```

```python
def gradient_descent(start_x, start_y, learning_rate, num_iterations):
    x_history = [start_x]
    y_history = [start_y]

    x, y = start_x, start_y
    for _ in range(num_iterations):
        dx, dy = gradient(x, y)
        x -= learning_rate * dx
        y -= learning_rate * dy
        x_history.append(x)
        y_history.append(y)

    return x_history, y_history
```

```python
start_x, start_y = 3, 4
learning_rate = 0.1
num_iterations = 20
```

```python
x_history, y_history = gradient_descent(start_x, start_y, learning_rate, num_iterations)
z_history = [cost_function(x, y) for x, y in zip(x_history, y_history)]
```

```python
plt.figure(figsize=(8, 6))
plt.contour(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100), cost_function(*np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))), 50)
plt.plot(x_history, y_history, 'r-o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent in 2D')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100), cost_function(*np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))), alpha=0.5)
ax.plot(x_history, y_history, z_history, 'r-o')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Cost')
ax.set_title('Gradient Descent in 3D')
plt.show()
```
