```python
import numpy as np
import matplotlib.pyplot as plt

def momentum_optimizer_1d(loss_function, initial_params, learning_rate=0.1, momentum=0.9, num_iterations=50):
    params = initial_params.copy()
    velocity = 0.0
    params_history = [params.copy()] # added copy in the list append

    for _ in range(num_iterations):
        gradient = compute_gradient_1d(loss_function, params)
        velocity = momentum * velocity + learning_rate * gradient
        params -= velocity
        params_history.append(params.copy()) # added copy here

    return np.array(params_history).flatten() # added flatten here

def compute_gradient_1d(loss_function, params):
    return 2 * params

def loss_function_1d(x):
    return x**2

initial_params = np.array([5.0]) #changed to numpy array
params_history = momentum_optimizer_1d(loss_function_1d, initial_params)

plt.plot(params_history, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Parameter Value')
plt.title('Momentum Optimizer (1D)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def momentum_optimizer_2d(loss_function, initial_params, learning_rate=0.01, momentum=0.9, num_iterations=100):
    params = initial_params.copy()
    velocity = np.zeros_like(params)
    params_history = [params.copy()]

    for _ in range(num_iterations):
        gradient = compute_gradient_2d(loss_function, params)
        velocity = momentum * velocity + learning_rate * gradient
        params -= velocity
        params_history.append(params.copy())

    return np.array(params_history)

def compute_gradient_2d(loss_function, params):
    x, y = params
    dx = 2 * x
    dy = 2 * y
    return np.array([dx, dy])

def loss_function_2d(params):
    x, y = params
    return x**2 + y**2

initial_params = np.array([3.0, 4.0])
params_history = momentum_optimizer_2d(loss_function_2d, initial_params)

x_vals = params_history[:, 0]
y_vals = params_history[:, 1]

plt.plot(x_vals, y_vals, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Momentum Optimizer (2D)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def momentum_optimizer_3d(loss_function, initial_params, learning_rate=0.01, momentum=0.9, num_iterations=100):
    params = initial_params.copy()
    velocity = np.zeros_like(params)
    params_history = [params.copy()]

    for _ in range(num_iterations):
        gradient = compute_gradient_3d(loss_function, params)
        velocity = momentum * velocity + learning_rate * gradient
        params -= velocity
        params_history.append(params.copy())

    return np.array(params_history)

def compute_gradient_3d(loss_function, params):
    x, y, z = params
    dx = 2 * x
    dy = 2 * y
    dz = 2 * z
    return np.array([dx, dy, dz])

def loss_function_3d(params):
    x, y, z = params
    return x**2 + y**2 + z**2

initial_params = np.array([3.0, 4.0, 5.0])
params_history = momentum_optimizer_3d(loss_function_3d, initial_params)

x_vals = params_history[:, 0]
y_vals = params_history[:, 1]
z_vals = params_history[:, 2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_vals, y_vals, z_vals, marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Momentum Optimizer (3D)')
plt.show()
```
