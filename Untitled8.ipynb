```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x, y):
    return x**2 + y**2

def nag_optimization(learning_rate, momentum, iterations, initial_x, initial_y):
    x, y = initial_x, initial_y
    vx, vy = 0, 0
    path_x, path_y = [x], [y]

    for _ in range(iterations):

        lookahead_x = x + momentum * vx
        lookahead_y = y + momentum * vy
        gradient_x = 2 * lookahead_x
        gradient_y = 2 * lookahead_y


        vx = momentum * vx - learning_rate * gradient_x
        vy = momentum * vy - learning_rate * gradient_y


        x += vx
        y += vy

        path_x.append(x)
        path_y.append(y)

    return path_x, path_y


learning_rate = 0.1
momentum = 0.8
iterations = 50
initial_x, initial_y = 5, 5


path_x, path_y = nag_optimization(learning_rate, momentum, iterations, initial_x, initial_y)


x_range = np.linspace(-6, 6, 100)
y_range = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = objective_function(X, Y)

plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.plot(path_x, path_y, marker='o', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Nesterov Accelerated Gradient (NAG)')
plt.show()
```
